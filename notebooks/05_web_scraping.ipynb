# 05 — Web scraping reproductible
Objectif : expliquer l'extraction HTML sans dépendre d'Internet lors de l'exécution.

```python
from html.parser import HTMLParser

import pandas as pd


class TableParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.rows, self.current, self.in_cell = [], [], False

    def handle_starttag(self, tag, attrs):
        self.in_cell = tag in {"th", "td"}

    def handle_data(self, data):
        if self.in_cell and data.strip():
            self.current.append(data.strip())

    def handle_endtag(self, tag):
        if tag in {"th", "td"}:
            self.in_cell = False
        if tag == "tr" and self.current:
            self.rows.append(self.current)
            self.current = []


html = """
<table>
  <tr><th>Langage</th><th>Usage</th></tr>
  <tr><td>Python</td><td>Analyse</td></tr>
  <tr><td>SQL</td><td>Requête</td></tr>
</table>
"""
parser = TableParser()
parser.feed(html)
```

## Extraction et contrôle

```python
scraped = pd.DataFrame(parser.rows[1:], columns=parser.rows[0])
assert not scraped.empty
scraped
```

## Conclusion
Cet exemple déterministe couvre l'analyse d'un tableau HTML. Une collecte réelle doit respecter les conditions d'utilisation, `robots.txt`, les délais et la gestion des erreurs HTTP.
