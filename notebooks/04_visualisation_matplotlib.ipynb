# 04 — Visualisation avec Matplotlib
Objectif : produire une comparaison lisible et reproductible des pourboires moyens par jour.

```python
from pathlib import Path

import pandas as pd

from src.analysis import summarize_by_category
from src.visualization import plot_category_comparison

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
tips = pd.read_csv(PROJECT_ROOT / "data" / "raw" / "tips.csv")
summary = summarize_by_category(tips, "day", "tip")
```

## Visualisation

```python
axis = plot_category_comparison(
    summary,
    "day",
    "moyenne",
    PROJECT_ROOT / "images" / "comparaison_pourboires_par_jour.png",
)
axis.set_title("Pourboire moyen par jour")
axis.set_xlabel("Jour")
axis.set_ylabel("Pourboire moyen")
```

## Conclusion
Le graphique compare les catégories sans masquer le contexte : l'effectif reste disponible dans le tableau du notebook précédent.
