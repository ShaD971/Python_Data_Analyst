# 01 — Fondamentaux de Pandas
Objectif : charger un jeu de données avec des chemins portables, l'inspecter et produire des statistiques descriptives.

```python
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
iris = pd.read_csv(PROJECT_ROOT / "data" / "raw" / "iris.csv")
iris.head()
```

## Contrôle qualité

```python
quality = pd.DataFrame(
    {"type": iris.dtypes.astype(str), "valeurs_manquantes": iris.isna().sum()}
)
quality
```

## Analyse descriptive

```python
iris.describe(include="all").T
```
