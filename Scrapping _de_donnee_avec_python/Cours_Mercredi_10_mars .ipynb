```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen("https://fr.wikipedia.org/wiki/Palmar%C3%A8s_universitaires ")
Soup = BeautifulSoup(html.read())
```

```python
Soup
```

```python
table= Soup.find('table',{'class':'wikitable'})
for elt in table:
    print(elt)
 
```

```python
import pandas as pd
df=pd.read_html(str(table), index_col='Place')[0]
```

```python
List=[]
for siblings in Soup.find('table',{'class':'wikitable'}).tr.next_siblings:
    if siblings != '\n':
        List.append(siblings)
    print(siblings)
```

```python
Liens=[]
for i in List:
    Liens.append(i.find('a').get('href'))
```
